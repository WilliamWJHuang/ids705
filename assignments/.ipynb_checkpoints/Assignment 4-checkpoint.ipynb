```python
class K_means:
    def __init__(self, n_cluster):
        self.n_cluster = n_cluster
    def fit(X,max_iteration=100):
        #X shape = (n, 2)
        #initiate centroids
        X = np.array(X)
        randindex = np.random.choice(X.shape[0],self.n_cluster)
        centroids = X[randindex]
        
        #calculate euclidean distance in assigned groups
        for i in range(max_iteration):
            cluster = list()
            for j in range(X.shape[0]):
                stored_dist = []
                for m in centroids:
                    temp_dist = np.linalg.norm(X[j]-m)
                    stored_dist.append(temp_dist)
                assigned_cluster = stored_dist.index(min(stored_dist))
                cluster.append(assigned_cluster)
            cluster = np.array(cluster)
            #udpate cetroids
            new_centroids = []
            for i in range(self.n_cluster):
                feature1 = np.mean(X[cluster==i][:,0])
                feature2 = np.mean(X[cluster==i][:,1])
                new_centroids.append((feature1,feature2))
            if np.all(new_centroids == centroids):
                centroids = new_centroids
                print("K-means training completed!")
                break
            else:
                centroids = new_centroids
        return (cluster, centroids)
    def cost(X,cluster, centroids):
        cost = 0
        for i in range(self.n_cluster):
            cost += np.linalg.norm(X[cluster==i]-centroids[i])
        return cost
        
```

```python
from sklearn.datasets.samples_generator import make_blobs
X, y = make_blobs(n_samples=500, centers=2, n_features=2,random_state=0)
kmeans = K_means(n_cluster=5)
cluster, centroids = kmeans.fit(X)
```

### [25 points] Visualizing and clustering digits with PCA and t-SNE

**(a)** Load the `scikit-learn` digits dataset. Apply PCA and reduce the data (with the associated cluster labels 0-9) into a 2-dimensional space. Plot the resulting 2-dimensional representation of the data.

**(b)** t-distributed stochastic neighborhood embedding (t-SNE) is a nonlinear dimensionality reduction technique that is particularly adept at embedding the data into lower 2 or 3 dimensional spaces. Apply t-SNE to the digits dataset and plot it in 2-dimensions (with associated cluster labels 0-9). You may need to adjust the parameters to get acceptable performance. You can read more about how to use t-SNE effectively [here](https://distill.pub/2016/misread-tsne/). A video introducing this method can be found [here](https://www.youtube.com/watch?v=RJVL80Gg3lA&list=UUtXKDgv1AVoG88PLl8nGXmw) for those who are interested.

*NOTE: An important note on t-SNE is that it is an example of transductive learning. This means that the lower dimensional representation of the data is only applicable to the specific input data - you can't just add a new sample an plot it in the sample 2-dimensional space without entirely rerunning the algorithm and finding a new representation of the data.*

**(c)** Compare/contrast the performance of these two techniques. Which seemed to cluster the data best and why? Given the comparative clustering performance that you observed and the note on t-SNE above, what are the pros and cons of PCA and t-SNE? *Note: You typically will not have labels available.*

**ANSWER**

```python
from sklearn.datasets import load_digits
from sklearn.decomposition import PCA
digits = load_digits()

#(a)
digits_data = digits.data
digits_label = digits.target
pca = PCA(n_components=2)
trans_digits_data = pca.fit(digits_data).transform(digits_data)

plt.figure(figsize=(10,6))
for i in range(len(np.unique(digits_label))):
    trans_digits_i = trans_digits_data[digits_label==i,:]
    plt.scatter(trans_digits_i[:,0],trans_digits_i[:,1],alpha=0.6,label=i)
plt.legend()
plt.xlabel("feature1")
plt.ylabel("feature2")
plt.title("2D representation of data by PCA")
```

```python
#(b)
from sklearn.manifold import TSNE
digits_tsne = TSNE(n_components=2, perplexity=40, early_exaggeration=12.0, learning_rate=300, 
                   n_iter=1000).fit_transform(digits_data)

plt.figure(figsize=(10,6))
for i in range(len(np.unique(digits_label))):
    trans_digits_i = digits_tsne[digits_label==i,:]
    plt.scatter(trans_digits_i[:,0],trans_digits_i[:,1],alpha=0.6,label=i)
plt.legend()
plt.xlabel("feature1")
plt.ylabel("feature2")
plt.title("2D representation of data by t-SNE")
```

**(c)**
Based on the two plots above, t-SNE seems to be perform better because there are much more overlapping regions on the plot produced with PCA than t-SNE. 
- **PCA**: It performs well when the data is linear. PCA is efficient when handling larget datasets.
- **t-SNE**: It performs well for non-linear datasets. However, it's inefficient for large datasets. Sometime it only work well for data visualization, not dimension reduction.

```python
#Extract all the 5's
five_index = digits_label == 5
```

```python
#(a)
plt.figure(figsize=(15, 1))
plt.suptitle("Class 5 Examples", fontsize=14, y=1.4)
for i in range(15):
    plt.subplot(2, 15, i+1) 
    plt.imshow(digits.images[five_index][i],cmap='gray')
```

```python
#(b)
x_= []
y_= []

for i in range(1,65):
    x_.append(i)
    pca = PCA(n_components=i)
    trans_five_data = pca.fit(digits.data[five_index]).explained_variance_ratio_
    y_.append(sum(trans_five_data))

plt.plot(x_, y_, 'r--', linewidth=2, markersize=12)
plt.xlabel("number of principal component")
plt.ylabel("the fraction of variance explained")
plt.title("The Fraction of Variance Explained Versus Number of Principal Component", fontsize = 12, y=1)
```

```python
#(c)
plt.figure(figsize=(10,3))
plt.subplot(1,5,1)
plt.title("original image")
plt.imshow(digits.images[five_index][3],cmap='gray')
idx = 1
for i in [1,5,10,25]:
    pca_c = PCA(n_components=i)
    pca_c_data = pca_c.fit(digits.data[five_index]).transform(digits.data[five_index])
    pca_transformedback = pca_c.inverse_transform(pca_c_data)
    g = pca_transformedback.reshape(182,8,8)
    plt.subplot(1,5,idx+1)
    plt.imshow(g[3],cmap='gray')
    plt.title("{} component".format(i))
    idx += 1
```
